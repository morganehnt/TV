```python
import pandas as pd 
import matplotlib.pyplot as plt 
data=pd.DataFrame({
    'Gender':['f','f','m','f','m',
            'm','f','m','f','m','m'],
    'TV':[3.4,3.5,2.6,4.7,4.1,4.1,
    5.1,3.9,3.7,2.1,4.3]
    })
grouped=data.groupby('Gender')
print(grouped.describe())

fig, axs = plt.subplots(1,2,figsize=(12,5))

grouped.boxplot(subplots=False, ax=axs[0])
axs[0].set_title('Boxplot par genre')
axs[0].set_ylabel('Temps devant la TV')

grouped['TV'].mean().plot(kind='bar', color=['pink', 'lightblue'], ax=axs[1])
axs[1].set_title("Moyenne de TV par genre")
axs[1].set_ylabel("Temps de TV")

print(grouped.size())
print(grouped['TV'].mean())
print(grouped['TV'].std())

plt.tight_layout()
plt.show()
```
